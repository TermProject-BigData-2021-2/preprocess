## UpperCase

```python
text = d[1][1]
text = text[1:-1]
text = re.sub('(http|ftp|https|uhttp|$uhttp)://', 'URL', text)
text = text.split('|||')
text = ' '.join(text)
text = text.split(' ')
#text
```

```python
m = ['MBTI', 'ISTJ', 'ISFJ', 'INFJ', 'INTJ', 'ISTP', 'ISFP', 'INFP','INTP', 'ESTP', 'ESFP', 'ENFP', 'ENTP', 'ESTJ', 'ESFJ', 'ENFJ', 'ENTJ']
exceptURL = []
for item in text:
    if 'URL' in item:
        continue
    else:
        item = re.sub('[^\w]', ' ', item)
        item = re.sub(r'\d+', ' ', item)
        item = item.strip()
        if len(item) > 1:
            exceptURL.append(item)
#print(exceptURL)
```

```python
num = 0
for w in exceptURL:
    if w.isalpha:
        if w.isupper():
            if w not in m:
                if w not in imoticon:
                    print(w)
                    num += 1
num
```

## Count Uppercase

```python
upp_count = []
m = ['MBTI', 'ISTJ', 'ISFJ', 'INFJ', 'INTJ', 'ISTP', 'ISFP', 'INFP','INTP', 'ESTP', 'ESFP', 'ENFP', 'ENTP', 'ESTJ', 'ESFJ', 'ENFJ', 'ENTJ']
for s in d:
    text = s[1]
    text = text[1:-1]
    text = re.sub('(http|ftp|https|uhttp|$uhttp)://', 'URL', text)
    text = text.split('|||')
    text = ' '.join(text)
    text = text.split(' ')
    exceptURL = []
    for item in text:
        if 'URL' in item:
            continue
        else:
            item = re.sub('[^\w]', ' ', item)
            item = re.sub(r'\d+', ' ', item)
            item = item.strip()
            if len(item) > 1:
                exceptURL.append(item)
    num = 0
    for w in exceptURL:
        if w.isalpha:
            if w.isupper():
                if w not in m:
                    if w not in imoticon:
                        num += 1
    upp_count.append(num)
```
